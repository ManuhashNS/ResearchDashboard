# For CSE Dept.

```python
import pandas as pd
```

```python
df_cs = pd.read_csv("./all_CSE_faculty_merged.csv")
```

```python
df_cs.info()
```

```python
df_cs[df_cs.Authors.isna()==True]
```

```python
df_cs.Authors = df_cs.Authors.fillna(df_cs[df_cs.Authors.isna()==True].Author)
```

```python
df_cs[df_cs.Year.isna()==True]
```

```python
df_cs[df_cs.Source.isna()==True]
```

```python
df_cs.Source = df_cs.Source.fillna("Undisclosed")
```

```python
df_cs[df_cs.Source.isna()]
```

```python
df_cs[df_cs.Type.isna()]
```

```python
df_cs.Type = df_cs.Type.fillna("Other")
```

```python
df_cs.info()
```

```python
df_cs.to_csv("./all_CSE_faculty_merged.csv",index=False)
```

```python
df_cs.Year = df_cs.Year.fillna("Unknown")
```

```python
df_cs.info()
```

```python
df_cs.Year = df_cs.Year.astype(str).str.replace(r"\.0$", "", regex=True).replace("nan","Unknown")
```

```python
df_cs.info()
```
